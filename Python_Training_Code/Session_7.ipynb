```python
evens = list(range(2,13,2))
odds = list(range(1,12,2))
multiple_list = []
for even in evens:
    for odd in odds:
        result = even * odd
        stmt = f"{even} times {odd} = {result}"
        multiple_list.append(stmt)
print (multiple_list)
```

```python
# % modulus

fizzy_test = list(range(1,101,1))

for test in fizzy_test:
    if (test % 3 == 0) & (test % 5 ==0):
        print( "FizzBang")
    elif test % 3 ==0:
        print("Fizz")
    elif test % 5 == 0:
        print("Bang")
    else:
        print(test)
```
