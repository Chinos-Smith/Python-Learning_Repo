```python
# pandas
import pandas as pd #alias to make code simpler.
import os #only for local files/network files
childpath = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv"

child_data = pd.read_csv(childpath) #pd.read_excel() opens Excel files.

type(child_data)

child_data.info()
# child_data.head(15)
# child_data.tail(25)
```

```python
house_prices["Period_DT"] =pd.to_datetime(house_prices["Period"], format="%Y-%m")

house_prices

```

```python
#child_data.info()

child_data["PBD_DT"] = pd.to_datetime(child_data["PersonBirthDate"], format="%Y-%m-%d", errors="coerce")
child_data["EPBD_DT"] = pd.to_datetime(child_data["ExpectedPersonBirthDate"],format="%Y-%m-%d", errors="coerce")
child_data.head(10)
```

```python
#house_prices.info()

max_change = house_prices["Percentage change (monthly) All property types"].max()
min_change = house_prices["Percentage change (monthly) All property types"].min()
mean_price= house_prices["Average price All property types"].mean()
print(f"Max Change: {max_change}", "min change: {min_change}")
print(mean_price)
```

```python
#child_data.info()

collection_end = pd.to_datetime("31/03/2025", format="%d/%m/%Y") 

child_data["time_delta"] = collection_end - child_data["PBD_DT"]
#child_data
child_data["Age"] = child_data["time_delta"]/pd.Timedelta(365.25,"d")
child_data["Age"] = child_data["Age"].astype(int)
child_data ["Age"]
```

```python
print(child_data["Age"].min())
print(child_data["Age"].max())
print(child_data["Age"].mean())
print(child_data["Age"].median())
print(child_data["Age"].sum())
```
