```python
import pandas as pd

filepath_s5 = r"https://github.com/data-to-insight/ERN-sessions/raw/main/data/903_xlsx_errors.xlsx"

df_s5 = pd.read_excel(filepath_s5, sheet_name=None)
#df_s5.keys()
df_s5["header"].head(10)
```

```python
#102 error is dob is not valid date
#d/m/y
# to_datetime() - format argument - "%d/%m/%y"
# errors ="coerce" - NA where not correctly formatted
header_s5 = df_s5["header"]

header_s5['DOB_dt']= pd.to_datetime(header_s5["DOB"],format="%d/%m/%Y", errors="coerce")
invalid_dobs = header_s5["DOB_dt"].isna()
error_rows =header_s5[invalid_dobs]
error_rows
```
